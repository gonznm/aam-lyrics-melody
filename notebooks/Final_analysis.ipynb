# Final Statistical Analysis Notebook

This notebook has been built to merge the melodic and semantic results. 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from scipy import stats
from scipy.stats import t
import pickle
import scipy.stats as st
import pickle
```

## Original Version of the lyrics

```python
# loads the melodic analysis files for similar and dissimilar groups
with open('melodic_difference/similarity_original_melodic_difference.txt', "rb") as fp:  
    b_similarity_original = pickle.load(fp)
       
with open('melodic_difference/dissimilarity_original_melodic_difference.txt', "rb") as fp:  
    b_dissimilarity_original = pickle.load(fp)
    
    
# appends the results in two lists
list_similarity = []
for key in b_similarity_original.keys():
    list_similarity.append(b_similarity_original[key])
    
list_dissimilarity = []
for key in b_dissimilarity_original.keys():
    list_dissimilarity.append(b_dissimilarity_original[key])
    
    
print(np.median(list_similarity))

print(np.median(list_dissimilarity))

# Boxplots plotting
plt.figure()
sns.boxplot(list_similarity)
plt.show()
plt.figure()
sns.boxplot(list_dissimilarity)
plt.show()

# Histograms plotting
plt.figure()
sns.distplot(list_similarity)
plt.show()

plt.figure()
sns.distplot(list_dissimilarity)
plt.show()

# Stores the data in an array
X_original = np.c_[np.array(list_similarity), np.array(list_dissimilarity)]


from scipy import stats

# Performs the t-test
stats.ttest_ind(X_original[:,0], X_original[:,1], equal_var=True)
```

## Transliterated Version of the lyrics

```python
with open('melodic_difference/similarity_transliterated_melodic_difference.txt', "rb") as fp:  
    b_similarity_transliterated = pickle.load(fp)
    
    
with open('melodic_difference/dissimilarity_transliterated_melodic_difference.txt', "rb") as fp:  
    b_dissimilarity_transliterated = pickle.load(fp)
    
    
    
list_similarity_trans = []
for key in b_similarity_transliterated.keys():
    list_similarity_trans.append(b_similarity_transliterated[key])
    
    
list_dissimilarity_trans = []
for key in b_dissimilarity_transliterated.keys():
    list_dissimilarity_trans.append(b_dissimilarity_transliterated[key])
    
    
print(np.mean(list_similarity_trans))
print(np.std(list_similarity_trans))

print(np.median(list_dissimilarity_trans))
print(np.std(list_dissimilarity_trans))




plt.figure()
sns.boxplot(list_similarity_trans)
plt.show()
plt.figure()
sns.boxplot(list_dissimilarity_trans)
plt.show()



plt.figure()
sns.distplot(list_similarity_trans)
plt.show()


plt.figure()
sns.distplot(list_dissimilarity_trans)
plt.show()


X_trans = np.c_[np.array(list_similarity_trans), np.array(list_dissimilarity_trans)]


from scipy import stats

#print(stats.ttest_rel(X_trans[:,0],X_trans[:,1]))
print(stats.ttest_ind(X_trans[:,0], X_trans[:,1], equal_var=True))
```
